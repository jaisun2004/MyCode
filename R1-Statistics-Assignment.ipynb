## Print all the teams which have started playing between 1930-1980. (5 points)

```python
laliga['DebutYear'] = laliga['Debut'].str.split('-', expand = True)[0]
laliga['DebutYear'] = pd.to_numeric(laliga['DebutYear'], errors='coerce', downcast = 'signed')
#laliga.info()
laliga.loc[(laliga.DebutYear >= 1930) & (laliga.DebutYear <= 1980)]
```

## Print the list of teams which came Top 5 in terms of points (2.5 points)

```python
laliga['Points'] = pd.to_numeric(laliga['Points'], errors='coerce', downcast = 'signed')
laliga.nlargest(5,'Points')
```

## Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

```python
laliga['GoalsFor'] = pd.to_numeric(laliga['GoalsFor'], errors='coerce', downcast = 'signed')
laliga['GoalsAgainst'] = pd.to_numeric(laliga['GoalsAgainst'], errors='coerce', downcast = 'signed')

def Goal_diff_count(): 
    laliga['GoalDifference'] = laliga['GoalsFor'] - laliga['GoalsAgainst']
    print(laliga.head(2))
    gdmax = laliga.GoalDifference.max()
    gdmin = laliga.GoalDifference.min()
    print("Team with Maximum Goal Difference is : ", laliga['Team'][laliga.GoalDifference == gdmax], "with Goal difference of", gdmax)
    print("Team with Minimum Goal Difference is : ", laliga['Team'][laliga.GoalDifference == gdmin], " with Goal differnece of", gdmin)
```

```python
Goal_diff_count()
```

## Create a new column with the name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100
If there are any numerical error, replace it with 0%
Print the top 5 teams which have the highest Winning percentage

```python
laliga['GamesWon'] = pd.to_numeric(laliga['GamesWon'], errors='coerce', downcast = 'signed')
laliga['GamesPlayed'] = pd.to_numeric(laliga['GamesPlayed'], errors='coerce', downcast = 'signed')
laliga['WinningPercent'] = ((laliga['GamesWon'] / laliga['GamesPlayed'] )*100)
laliga['WinningPercent'].fillna(0, inplace = True)
laliga.nlargest(5, 'WinningPercent')
```
